## Kubernetes Components

- Pod
- Deployment
- Service
- ReplicaSet
- Namespace
- Node
- Cluster

````python
deployment_yaml = """

apiVersion: apps/v1

kind: Deployment

metadata:

  name: retailpulse-ai

spec:

  replicas: 2

  selector:

    matchLabels:

      app: retailpulse

  template:

    metadata:

      labels:

        app: retailpulse

    spec:

      containers:

      - name: retailpulse

        image: retailpulse-ai:latest

        ports:

        - containerPort: 8501

"""

print(deployment_yaml)
````

````python
with open("deployment.yaml","w") as file:

    file.write(deployment_yaml)

print("deployment.yaml created successfully.")
````

````python
service_yaml = """

apiVersion: v1

kind: Service

metadata:

  name: retailpulse-service

spec:

  selector:

    app: retailpulse

  ports:

  - protocol: TCP

    port: 80

    targetPort: 8501

  type: LoadBalancer

"""

print(service_yaml)
````

````python
with open("service.yaml","w") as file:

    file.write(service_yaml)

print("service.yaml created successfully.")
````

## Kubernetes Commands

## Architecture Explanation

The RetailPulse AI application is packaged as a Docker image and deployed to a Kubernetes cluster using a Deployment manifest.

Kubernetes creates multiple pods to ensure high availability. A LoadBalancer Service exposes the application to end users and distributes traffic across the running pods.
