# Model Performance Visualization

```python
plt.figure(figsize=(8,5))

plt.bar(
    comparison["Model"],
    comparison["Value"]
)

plt.title(
    "Forecasting Model Comparison"
)

plt.ylabel(
    "Metric Value"
)

plt.show()
```

# Metric Interpretation

The two forecasting models are evaluated using different metrics.

- **Prophet:** Mean Absolute Percentage Error (MAPE), expressed as a percentage.
- **LSTM:** Root Mean Squared Error (RMSE), expressed in sales units.

Since these metrics use different units and scales, they should not be directly compared numerically. Each metric should be interpreted independently based on its intended purpose.

# Model Analysis

Prophet:
- Easy to interpret
- Captures trend and seasonality
- Fast training

LSTM:
- Learns temporal dependencies
- Handles sequential patterns
- Suitable for complex forecasting tasks

# Recommended Model

Based on the evaluation results:

- Prophet provides interpretable forecasting with trend and seasonality analysis.
- LSTM captures complex temporal patterns and sequential relationships.
- For advanced retail forecasting applications, LSTM is recommended when sufficient historical data is available.
- Prophet remains valuable for fast deployment and business interpretability.
