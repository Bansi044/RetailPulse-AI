```python
import mlflow
import pandas as pd
```

```python
mlflow.set_experiment(
    "RetailPulse_Week1"
)
```

# Prophet Experiment Logging

```python
with mlflow.start_run(
    run_name="Prophet_Model"
):

    mlflow.log_param(
        "model_type",
        "Prophet"
    )

    mlflow.log_param(
        "forecast_horizon",
        30
    )

    mlflow.log_metric(
        "mape",
        39.40
    )

print(
    "Prophet experiment logged."
)
```

# LSTM Experiment Logging

```python
with mlflow.start_run(
    run_name="LSTM_Model"
):

    mlflow.log_param(
        "model_type",
        "LSTM"
    )

    mlflow.log_param(
        "epochs",
        100
    )

    mlflow.log_param(
        "sequence_length",
        12
    )

    mlflow.log_metric(
        "rmse",
        21232.65
    )

print(
    "LSTM experiment logged."
)
```

```python
import mlflow

client = mlflow.tracking.MlflowClient()

client.search_experiments()
```
