# What is CI/CD?

Continuous Integration (CI):
Developers regularly merge code into a shared repository where automated tests are executed.

Continuous Deployment (CD):
After successful testing, the application is automatically deployed to production or staging environments.

```python
pipeline_steps = pd.DataFrame({

    "Stage":[
        "Code Commit",
        "Build",
        "Testing",
        "Docker Image",
        "Deployment",
        "Monitoring"
    ],

    "Purpose":[
        "Push Code",
        "Compile Project",
        "Validate Code",
        "Package Application",
        "Release Application",
        "Track Performance"
    ]

})

pipeline_steps
```

```python
cicd_yaml = """
name: RetailPulse AI CI/CD

on:
  push:
    branches:
      - main

jobs:
  build-and-deploy:

    runs-on: ubuntu-latest

    steps:

      - name: Checkout Repository
        uses: actions/checkout@v4

      - name: Setup Python
        uses: actions/setup-python@v4
        with:
          python-version: '3.11'

      - name: Install Dependencies
        run: pip install -r requirements.txt

      - name: Run Tests
        run: echo "Running tests..."

      - name: Build Docker Image
        run: docker build -t retailpulse-ai .

      - name: Deploy Application
        run: echo "Deployment triggered successfully."
"""

with open("deploy.yml", "w") as f:
    f.write(cicd_yaml)

print("GitHub Actions workflow created successfully!")

print(cicd_yaml)
```
