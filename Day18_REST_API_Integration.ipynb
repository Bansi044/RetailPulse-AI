# What is a REST API?

A REST API (Representational State Transfer Application Programming Interface) allows applications to communicate over HTTP.

It enables data exchange between machine learning models, dashboards, mobile applications, and business systems.

```python
api_endpoints = pd.DataFrame({

    "Endpoint":[
        "/forecast",
        "/customers",
        "/sales",
        "/metrics",
        "/health"
    ],

    "Purpose":[
        "Return Forecast",
        "Customer Insights",
        "Sales Summary",
        "Model Metrics",
        "System Status"
    ]

})

api_endpoints
```

```python
from fastapi import FastAPI
import pandas as pd

# Create FastAPI application
app = FastAPI(
    title="RetailPulse AI API",
    description="REST API for Retail Analytics and Demand Forecasting",
    version="1.0"
)

# Health Check Endpoint
@app.get("/health")
def health_check():
    return {
        "status": "Healthy",
        "model": "RetailPulse AI v1.0"
    }

# Forecast Endpoint
@app.get("/forecast")
def get_forecast():

    df = pd.read_csv("prophet_forecast_results.csv")

    return df[["ds", "yhat"]].tail(30).to_dict(
        orient="records"
    )

# Metrics Endpoint
@app.get("/metrics")
def get_metrics():

    return {
        "Prophet_MAPE": "39.40%",
        "LSTM_RMSE": 21232.65
    }

print("✅ FastAPI application created successfully!")

print("Available Endpoints:")
print("/health")
print("/forecast")
print("/metrics")
```

```python
from fastapi import FastAPI
import pandas as pd

app = FastAPI(
    title="RetailPulse AI API",
    version="1.0"
)

@app.get("/health")
def health():
    return {
        "status": "ok",
        "model": "RetailPulse AI v1.0"
    }

@app.get("/forecast")
def get_forecast():
    df = pd.read_csv("prophet_forecast_results.csv")
    return df[["ds", "yhat"]].tail(30).to_dict(orient="records")

@app.get("/metrics")
def get_metrics():
    return {
        "prophet_mape": 39.40,
        "lstm_rmse": 21232.65
    }

print("✅ FastAPI application created successfully.")
print("Available endpoints:")
print("/health")
print("/forecast")
print("/metrics")
```
