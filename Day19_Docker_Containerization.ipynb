# What is Docker?

Docker is an open-source platform that packages applications, libraries, and dependencies into lightweight containers.

Containers ensure that applications behave the same way regardless of the environment in which they are executed.

```python
docker_components = pd.DataFrame({

    "Component":[
        "Application",
        "Dependencies",
        "Docker Image",
        "Docker Container",
        "Cloud Server"
    ],

    "Purpose":[
        "RetailPulse AI",
        "Python Libraries",
        "Packaged Application",
        "Running Instance",
        "Deployment Platform"
    ]

})

docker_components
```

```python
dockerfile_content = """
FROM python:3.11-slim

WORKDIR /app

COPY requirements.txt .

RUN pip install --no-cache-dir -r requirements.txt

COPY . .

EXPOSE 8501

CMD ["streamlit", "run", "app.py", "--server.address=0.0.0.0", "--server.port=8501"]
"""

with open("Dockerfile", "w") as f:
    f.write(dockerfile_content)

print("Dockerfile created successfully!")

print(dockerfile_content)
```
