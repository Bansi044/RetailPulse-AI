# Why Retraining is Required

Machine learning models are trained on historical data.

Over time:

- Customer preferences change
- Demand patterns change
- Market conditions change
- Data drift occurs

Retraining helps models adapt to new business conditions.

# Retraining Triggers

A model should be retrained when:

1. Forecasting accuracy drops significantly.

2. Data drift is detected.

3. New business patterns emerge.

4. New data becomes available.

5. Scheduled retraining intervals are reached.

# Monitoring Metrics

The following metrics should be monitored:

- RMSE
- MAPE
- Forecast Accuracy
- Data Drift Indicators
- Business KPIs
